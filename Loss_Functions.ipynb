# Loss Functions

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {
    'clf': (np.random.binomial(1, .5, 100),
            np.random.rand(100)), #y, yhat
    'reg': (.2 * np.arange(100) + 5,
            .2 * np.arange(100) + 5 + np.random.uniform(-10, 10, size=(100)))
}
```

```python
def cross_entropy(y, yhat):
    return np.round(-1*(y*np.log(yhat) + (1-y)*np.log(1-yhat)), 5)


def mae(y, yhat):
    return np.abs(y - yhat)/len(y)


def mse(y, yhat):
    return (y - yhat)**2 / len(y)


def loss_f(f, y, yhat):
    loss_functions = {
        'cross_entropy': cross_entropy,
        'mae': mae,
        'mse': mse,
    }
    return loss_functions[f](y, yhat)

```

```python
plt.figure(figsize=(21, 5))

fig, axs = plt.subplots(1, 3, figsize=(21, 7))

axs[0].plot(np.arange(0, 100), loss_f('cross_entropy', *data['clf']), marker='o', color='r', label='Cross Entropy')
axs[1].plot(np.arange(0, 100), loss_f('mae', *data['reg']), marker='o', color='pink', label='MAE')
axs[2].plot(np.arange(0, 100), loss_f('mse', *data['reg']), marker='o', color='crimson', label='MSE')

# plt.plot(np.arange(0, 100), yhat, color='g')
# plt.bar(np.arange(0, 100), y, alpha=.7)

fig.legend(title='Loss Functions', ncol=3, loc='upper center')
```
