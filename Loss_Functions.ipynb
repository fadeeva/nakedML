# Loss Functions

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
y = np.random.binomial(1, .5, 100)
yhat = np.random.rand(100)
```

```python
def cross_entropy(y, yhat):
    return np.round(-1*(y*np.log(yhat) + (1-y)*np.log(1-yhat)), 5)
```

```python
plt.figure(figsize=(21, 5))
plt.plot(np.arange(0, 100), cross_entropy(y, yhat), marker='o', color='r')
plt.plot(np.arange(0, 100), yhat, color='g')
plt.bar(np.arange(0, 100), y, alpha=.7)
```
